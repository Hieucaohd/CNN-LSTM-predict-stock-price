```python
import yfinance as yf
from datetime import  datetime
import pandas as pd
```

```python
symbol = "KO"
start_date = "1980-01-01"
interval = "1wk"
mapping_symbol = {
    "0P0000HY8X.VN": "VNINDEX",
    "KO": "coca.cola"
}
stock = yf.Ticker(symbol)
```

```python
stock.info
```

```python
hist = stock.history()
```

```python
hist
```

```python
hist = stock.history(
    start=start_date,
    interval=interval
)
```

```python
str(hist.index.min().date())
```

```python
str(hist.index.max().date())
```

```python
hist.index = hist.index.strftime('%Y-%m-%d %H:%M:%S')
hist.index = pd.to_datetime(hist.index).tz_localize('UTC')
```

```python
hist
```

```python
symbol =  mapping_symbol.get(symbol, symbol)
```

```python
symbol = symbol.replace(".", "_").lower()
```

```python
file_name = f"../../data/{symbol}/{symbol}-{str(hist.index.min().date())}-to-{str(hist.index.max().date())}_{interval}.csv"
```

```python
file_name = file_name.replace("-", "_")
```

```python
file_name
```

```python

hist.to_csv(file_name)
```
