```python
import yfinance as yf
from datetime import  datetime
import pandas as pd
```

```python
symbol = "KO"
stock = yf.Ticker(symbol)
```

```python
hist = stock.history()
```

```python
interval = "1wk"
hist = stock.history(
    start="1995-01-01",
    interval=interval
)
```

```python
str(hist.index.min().date())
```

```python
str(hist.index.max().date())
```

```python
hist.index = hist.index.strftime('%Y-%m-%d %H:%M:%S')
hist.index = pd.to_datetime(hist.index).tz_localize('UTC')
```

```python
hist
```
