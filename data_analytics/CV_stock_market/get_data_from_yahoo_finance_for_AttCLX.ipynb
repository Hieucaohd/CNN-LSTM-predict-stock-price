```python
import yfinance as yf
from datetime import  datetime
```

```python
symbol = "601398.SS"
stock = yf.Ticker(symbol)
```

```python
stock.info
```

```python
hist = stock.history()
```

```python
hist
```

```python
hist = stock.history(
    start="1800-01-01"
)
```

```python
str(hist.index.min().date())
```

```python
str(hist.index.max().date())
```

```python
hist["trade_date"] = hist.index
```

```python
hist.rename(
    columns={
        "Close": "close",
        "Open": "open",
        "High": "high",
        "Low": "low",
        "Volume": "volume"
    },
    inplace=True
)
```

```python
min_date = hist.index.min()
max_date = hist.index.max()
```

```python
hist["trade_date"] = hist["trade_date"].dt.strftime('%Y%m%d')
```

```python
hist = hist[['trade_date', 'open', 'high', 'low', 'close', 'volume' ]]
```

```python
hist.index = range(0, len(hist))
```

```python
hist
```

```python
symbol = symbol.replace(".", "_").lower()
```

```python
file_name = f"./data/{symbol}-{str(min_date.date())}-to-{str(max_date.date())}.csv"
```

```python
file_name = file_name.replace("-", "_")
```

```python
file_name
```

```python

hist.to_csv(file_name)
```
